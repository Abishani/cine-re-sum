### converting dataset from json to csv

```python
import pandas as pd

df = pd.read_json('/content/rottentomatoes.json')
```

```python
df
```

```python
# concatenating multiple reviews by different critics
df['critics'] = df['_critics'].map(lambda x: ' '.join(x.values())).tolist()
df['critics'] = [''.join(critic) for critic in df['critics']]
df['_critic_consensus'] = df['_critic_consensus'].values.tolist()

# #displaying concatenated multiple reviews and summaries
# df = df[['critics', '_critic_consensus']]
df.drop(['_critics'], axis=1,inplace=True)
df.head()
```

```python
df.to_csv('rotten_tomatoes_reviews.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Print the number of examples in each set
print(f"Number of training examples: {len(train_data)}")
print(f"Number of testing examples: {len(test_data)}")

# Print the number of reviews(critics) and summary(_critics_concensus) in each train and test set
# print(df[test_data][1]["critics"])


print(f"Number of training examples: {test_data['critics']}\n")
print(f"Number of training examples: {test_data['_critic_consensus']}")
```
